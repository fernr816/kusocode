```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
fig = plt.figure(figsize=[10, 10])

# init
x_next = 0
y_next = 0

for i in range(10000):
    p = np.random.randint(0, 100)
    plt.plot(x_next, y_next, '.', color='black', markersize=1)
    
    if p < 1:
        x, y = x_next, y_next
        x_next = 0
        y_next = 0.16*y
    elif 0 < p & p < 8:
        x, y = x_next, y_next
        x_next = 0.2*x - 0.26*y
        y_next = 0.23*x + 0.22*y + 1.6
    elif 7 < p & p < 15:
        x, y = x_next, y_next
        x_next = -0.15*x + 0.28*y
        y_next = 0.26*x + 0.24*y + 0.44
    else:
        x, y = x_next, y_next
        x_next = 0.85*x + 0.04*y
        y_next = -0.04*x + 0.85*y + 1.6

# display setting
plt.xlim(-5, 5)
plt.ylim(-1, 11)
plt.box(False)
plt.tick_params(labelbottom=False, labelleft=False, bottom=False, left=False)

plt.savefig("img.png")
```
